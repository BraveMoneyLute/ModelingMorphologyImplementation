### Interpreting the values
`activation_matrix`: Every column is the sum of activations from one event (eg. `#b_w#_ow_bl_lo_blow`) with each of the outcomes  
`weight_matrix`: Every column is the activation from one digraph with each of the outcomes

I will look at the the `bl`,`cl`,`cr`,`dr` diphones and their respective weights to the outcomes, as these are the sound combinations that are associated with the phonestheme meanings.

````python
weight_matrix[['#b','bl','#c','cl','cr','#d','dr']]
````

Cool! the respective outcomes have the highest weights!

Now, to actually make use of our activation matrix, we could look for the most iconic words in our list which would be most "representative" of the eight phonestheme categories. We can use the indices from the  wordlist to identify which words have the highest activations:

````python
activation_matrix.idxmax(axis=1)
````
